```python
# Buat mapping sector ke region (menggunakan modus sebagai nilai default)
sector_to_region = data_travel.dropna(subset=['region']).groupby('sector')['region'].agg(lambda x: x.mode()[0] if not x.mode().empty else 'not known').to_dict()

# Isi missing value di region berdasarkan sector menggunakan .loc
data_travel.loc[data_travel['region'].isna(), 'region'] = data_travel['sector'].map(sector_to_region).fillna('not known')

# Cek apakah masih ada missing value
print(f"Missing value di region setelah imputasi: {data_travel['region'].isna().sum()}")
```

```python
# Hitung jumlah missing value pada class berdasarkan airline
missing_class_counts = data_travel[data_travel['class'].isna()]['airline'].value_counts()

# Hitung total jumlah tiket untuk tiap airline dalam dataset
total_airline_counts = data_travel['airline'].value_counts()

# Gabungkan kedua informasi ke dalam satu DataFrame
summary_data_travel = pd.DataFrame({
    'Airline': missing_class_counts.index,  # Nama maskapai
    'Missing_Class_Count': missing_class_counts.values,  # Jumlah missing class
    'Total_Airline_Count': total_airline_counts.reindex(missing_class_counts.index).values  # Total tiket maskapai tersebut
})

# Tampilkan hasil
print(summary_data_travel)
```

```python
# Hitung jumlah kelas unik per airline
unique_class_per_airline = data_travel.groupby('airline')['class'].nunique()

# Filter airline yang hanya memiliki satu jenis kelas
single_class_airlines = unique_class_per_airline[unique_class_per_airline == 1]

# Konversi ke DataFrame untuk tampilan lebih rapi
single_class_airlines_data_travel = single_class_airlines.reset_index()
single_class_airlines_data_travel.columns = ['Airline', 'Unique_Class_Count']

# Tampilkan hasil
print(single_class_airlines_data_travel)
```

```python
# Filter data hanya untuk airline yang memiliki satu kelas
filtered_data = data_travel[data_travel['airline'].isin(single_class_airlines_data_travel['Airline'])]

# Cek apakah ada missing value di kolom 'class' pada airline ini
missing_in_single_class = filtered_data['class'].isna().sum()

# Output hasil
print(f"Jumlah missing value pada class untuk airline yang hanya punya satu kelas: {missing_in_single_class}")
```

```python
# Daftar kategori kelas
economy=['X','Q','V','L','T','P','N','A','H','B','Y','E','Z','O','I','G','K','M','S','D','C','F','R','U','J','ECONOMY']
business=['B1','E2','O1','W1','H1','L1','X1','E3','N1','E1','S1','C3','Y1','E4','T1','M1','A1','R1','F1','C2','P1','B2','E5']
first_class=['H3','Y3','S3','P3','R3','M3','O3','N3','X3','W3','L3','T3']

# Fungsi untuk mengategorikan kelas
def categorize_class(class_code):
    if class_code in economy:
        return 'Economy'
    elif class_code in business:
        return 'Business'
    elif class_code in first_class:
        return 'First Class'
    else:
        return 'Unknown'

# Tambahkan kolom baru ke dataset
data_travel['class_category'] = data_travel['class'].apply(categorize_class)
```
